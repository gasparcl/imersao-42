```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000
low_limit = 10000
high_limit = 20000

# FunÃ§Ã£o para mapear os preÃ§os para as categorias
def categorize_price(price):
    if price < low_limit:
        return 'low'
    elif low_limit <= price <= high_limit:
        return 'medium'
    else:
        return 'high'

# Aplicar a funÃ§Ã£o Ã  coluna de preÃ§os para criar uma nova coluna de categorias
CarData['price_category'] = CarData['price'].apply(categorize_price)

# Mapear os preÃ§os medianos das empresas de carros para as categorias
median_prices = CarData.groupby('CarCompany')['price'].median()
company_price_category = median_prices.apply(categorize_price)

# Mapear os preÃ§os medianos das empresas de carros na coluna 'CarCompany' no conjunto de dados
CarData['CarCompany'] = CarData['CarCompany'].map(company_price_category)

# Verificar os valores Ãºnicos na coluna de categorias
CarData['CarCompany'].unique()
```

```python
def r2_score(y_true, y_pred):
    """
    Calcula o coeficiente de determinaÃ§Ã£o R^2 entre os valores reais e preditos.

    Args:
    - y_true: Array-like, valores reais.
    - y_pred: Array-like, valores preditos.

    Returns:
    - r2: Coeficiente de determinaÃ§Ã£o R^2.
    """
    # MÃ©dia dos valores reais
    mean_y_true = sum(y_true) / len(y_true)

    # Soma total dos quadrados (SST)
    sst = sum((y_true - mean_y_true) ** 2)

    # Soma dos quadrados dos resÃ­duos (SSE)
    sse = sum((y_true - y_pred) ** 2)

    # Coeficiente de determinaÃ§Ã£o R^2
    r2 = 1 - (sse / sst)

    return r2
```
