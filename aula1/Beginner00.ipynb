```python
# You need to import the pandas and seaborn library
import pandas as pd
import seaborn as sns
```

```python
#You have to reproduce this graph
sns.set_theme()

salary_df = pd.read_csv("Salary_dataset.csv")

sns.jointplot(
    data=salary_df,
    x="YearsExperience",
    y="Salary",
    color="#9b91c0"
)
```

```python
# We're going to use this library and find out how it works on the Internet
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Separate the features (YearsExperience) from the target variable (Salary)
X =  salary_df[["YearsExperience"]]
y = salary_df[["Salary"]]

# Splitting data into training/testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)

# Create a linear regression model
regr = LinearRegression()

# Train the model
train = regr.fit(X_train, y_train)

# Predict the Salary values based on YearsExperience
salary_predictions = regr.predict(X_test)

# Coefficients
print("Coefficients: \n", regr.coef_)

# Mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, salary_predictions))

# The coefficient of determination
print("Coefficient of determination: %.2f" % r2_score(y_test, salary_predictions))
```

```python

import matplotlib.pyplot as plt

# Create a linear regression plot using lmplot
sns.lmplot(
    salary_df[["YearsExperience", "Salary"]],
    x="YearsExperience",
    y="Salary",
    height=7,
    aspect=1.1
)

# Set labels and title
plt.title("Linear Regression: Years of Exeperience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
